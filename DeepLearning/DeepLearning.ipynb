```python
import numpy as np
```

Step 1: Prepare Training Data

```python
#Training Data
X=np.array([1,2,3,4])  # Study hours
y=np.array([0,0,1,1])  # Fail(0)/Pass(1)

X,y
```

Step 2: Initialize Parameters

```python
w=0.0
b=0.0
learning_rate=1
epochs=5
print("Initial Weight:",w)
print("Initial Bias:",b)
```

Define Activation Function (Step Function)

```python
def step_function(z):
    return 1 if z>=0 else 0
```

Training Loop

```python
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):
        #1. Compute Weighted sum
        z = w * X[i] + b

        #2.Apply Activation Function
        y_pred=step_function(z)

        #3.Compute Error
        error=y[i]-y_pred

        #4.Update Weight
        w=w+learning_rate*error*X[i]

        #5.Update Bias
        b=b+learning_rate*error

        print(f"Input:{X[i]}, Predicted: {y_pred}, Actual: {y[i]},w:{w},b:{b}")
        
```

```python
print("\n Final Weight:",w)
print("Final Bias:",b)
```

Testing the Model

```python
test_value=2.9
z=w*test_value+b
prediction=step_function(z)
print(f"For study hours {test_value}, Prediction",prediction)
```

Perceptron AND Gate

```python
X=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
y=np.array([0,0,0,1])

X,y
```

```python
w=np.zeros(2)
b=0
learning_rate=0.1
epochs=10

print("Initial Weights:",w)
print("Initial Bias:",b)
```

```python
def step_function(z):
    return 1 if z>=0 else 0
```

Train Perceptron

```python
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):
        
        # 1. Compute Weighted sum
        z = np.dot(w, X[i]) + b
        
        # 2. Apply Activation Function
        y_pred = step_function(z)

        # 3. Compute Error
        error = y[i] - y_pred

        # 4. Update Weights
        w = w + learning_rate * error * X[i]

        # 5. Update Bias
        b = b + learning_rate * error

        print(f"Input:{X[i]}, Predicted:{y_pred}, Actual:{y[i]}, w:{w}, b:{b}")
```

```python
print("\n Final Weight:",w)
print("Final Bias:",b)
```

Test Model

```python
print("\n Testing AND Gate")
for i in range(len(X)):
    z=np.dot(w,X[i])+b
    prediction=step_function(z)
    print("Input:",X[i],"Prediction:",prediction)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([
    [1, 1],   # E1
    [1, 0],   # E2
    [0, 1],   # E3
    [0, 0]    # E4
])

y = np.array([1, 1, 1, 0])   
```

```python
w = np.zeros(2)
b = 0
learning_rate = 1
epochs = 10
```

```python
def step_function(z):
    return 1 if z > 0 else 0
```

```python
for epoch in range(epochs):
    for i in range(len(X)):
        
        z = np.dot(w, X[i]) + b
        y_pred = step_function(z)
        error = y[i] - y_pred
        
        w = w + learning_rate * error * X[i]
        b = b + learning_rate * error
```

```python
print("Final Weights:", w)
print("Final Bias:", b)
```

```python
print("\nDecision Boundary Equation:")
print(f"{w[0]}*x1 + {w[1]}*x2 + {b} = 0")
```

```python
plt.figure()

# Plot data points
for i in range(len(X)):
    if y[i] == 1:
        plt.scatter(X[i][0], X[i][1], marker='o')
    else:
        plt.scatter(X[i][0], X[i][1], marker='x')

# Create boundary line
x_values = np.linspace(-0.5, 1.5, 100)

if w[1] != 0:
    y_values = -(w[0]*x_values + b) / w[1]
    plt.plot(x_values, y_values)
else:
    plt.axvline(x=-b/w[0])

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel("Suspicious Keywords (x1)")
plt.ylabel("Unknown Sender (x2)")
plt.title("Perceptron Decision Boundary (Spam Detection)")
plt.grid()
plt.show()
```

```python
print("\n\nTesting Modified Dataset (E3 -> 0)")

y_modified = np.array([1, 1, 0, 0])

w = np.zeros(2)
b = 0

for epoch in range(epochs):
    for i in range(len(X)):
        z = np.dot(w, X[i]) + b
        y_pred = step_function(z)
        error = y_modified[i] - y_pred
        
        w = w + learning_rate * error * X[i]
        b = b + learning_rate * error

print("Modified Final Weights:", w)
print("Modified Final Bias:", b)
```
